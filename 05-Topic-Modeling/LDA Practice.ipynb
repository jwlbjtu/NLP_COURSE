```python
import pandas as pd
```

```python
npr = pd.read_csv('npr.csv')
```

```python
npr.head()
```

```python
len(npr)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
```

```python
dtm = cv.fit_transform(npr['Article'])
```

```python
dtm
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
LDA = LatentDirichletAllocation(n_components=7, random_state=42)
```

```python
LDA.fit(dtm)
```

```python
# Grab the vocavulary of words
```

```python
len(cv.get_feature_names())
```

```python
import random

random_word_id = random.randint(0, 54777)

cv.get_feature_names()[random_word_id]
```

```python
# Grab the topics
```

```python
LDA.components_
```

```python
len(LDA.components_)
```

```python
LDA.components_.shape
```

```python
single_topic = LDA.components_[0]
```

```python
top_ten_words = single_topic.argsort()[-10:]
top_ten_words
```

```python
for index in top_ten_words:
    print(cv.get_feature_names()[index])
```

```python
# Grab the highest probability words per topic
```

```python
for i, topic in enumerate(LDA.components_):
    print(f"THE TOP 15 WORDS FOR TOPIC #{i}")
    print([cv.get_feature_names()[index] for index in topic.argsort()[-15:]])
    print("\n")
    print("\n")
```

```python
topic_results = LDA.transform(dtm)
```

```python
topic_results.shape
```

```python
topic_results[0].round(2).argmax()
```

```python
npr['Topic#'] = topic_results.argmax(axis=1)
```

```python
npr
```

```python
topic_names = {0: "BBLP", 1: "political", 2: "education", 3: "health care", 4: "infrastructure", 5: "personal life", 6: "motor"}
```

```python
npr['Topic'] = [topic_names[key] for key in topic_results.argmax(axis=1)]
```
